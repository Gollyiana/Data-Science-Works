```python
# Cell 1: Setup and Data Loading
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set plotting style for better visualization
sns.set_style("whitegrid")

# Load the Dataset. MUST be in the same folder as this notebook!
file_name = 'blood_donation_data.csv'
df = pd.read_csv(file_name)

# Display data size (For Assignment Section 1: Dataset Description)
print("--- Dataset Information ---")
print(f"Total Observations (Rows): {df.shape[0]}")
print(f"Total Variables (Columns): {df.shape[1]}")
print("\nFirst 5 Rows:")
print(df.head())
```

```python
# Cell 2: Descriptive Statistics (Assignment Section 2)

numerical_cols = ['Total_Donations', 'Age', 'Hemoglobin_g_dL', 'Weight_kg']

print("\n--- Descriptive Statistics Table ---")

# Calculate required statistics
descriptive_stats = df[numerical_cols].describe().T
stats_to_report = descriptive_stats[['mean', '50%', 'std']]
stats_to_report.columns = ['Mean', 'Median', 'Standard Deviation']

# Calculate Mode and add to the table
mode_series = df[numerical_cols].mode().iloc[0]
stats_to_report['Mode'] = mode_series.values

# Display the final summary table
print(stats_to_report)
```

```python
# Cell 3: Age vs. Total Donations (Scatter Plot)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Total_Donations', data=df, alpha=0.6)
plt.title('1. Relationship between Donor Age and Total Donations')
plt.xlabel('Age (Years)')
plt.ylabel('Total Donations (Frequency)')
# --- THE CHANGE IS HERE ---
plt.savefig('1_Age_vs_Total_Donations_Scatter.png')
# ---
# You can remove or comment out plt.show() if you don't want it displayed in the notebook
# plt.show()
```

```python
# Cell 4: Relationship 2 (Box Plot)
plt.figure(figsize=(8, 6))
sns.boxplot(x='Gender', y='Hemoglobin_g_dL', data=df)
plt.title('2. Distribution of Hemoglobin Levels by Donor Gender')
plt.xlabel('Gender')
plt.ylabel('Hemoglobin (g/dL)')
plt.show()
```

```python
# Cell 5: Relationship 3 (Bar Chart)
mean_donations_by_group = df.groupby('Blood_Group')['Total_Donations'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(x=mean_donations_by_group.index, y=mean_donations_by_group.values, palette='viridis')
plt.title('3. Average Total Donations by Blood Group')
plt.xlabel('Blood Group')
plt.ylabel('Mean Total Donations')
plt.tight_layout()
plt.show()
```
